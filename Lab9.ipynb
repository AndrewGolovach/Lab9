### Задание 1

Расмотрим функцию $TPR(T)$ (true positive rate), зависящую от порогового параметра $T$:

$$
TPR(T) 
= \frac {\sum\limits_{(x, y)} 1_{h_S(x, T) = 1} * 1_{y = +1}} {\sum\limits_{(x, y)} 1_{y = +1}} 
= \int_T^{+\infty} f_{pos}(t) \mathrm{d}t
$$

где $f_{pos}(t)$ - плотность распределения случайной величины $X$, которая оценивает "принадлежность" объекта положительному классу.

Через $FPR(T)$ обозначим функцию false positive rate, которая будет также зависеть от порогового параметра $T$ и задается формулой:

$$
FPR(T) 
= \frac {\sum\limits_{(x, y)} 1_{h_S(x, T) = 1} * 1_{y = -1}} {\sum\limits_{(x, y)} 1_{y = -1}} 
= \int_T^{+\infty} f_{neg}(t) \mathrm{d}t
$$

где $f_{neg}(t)$ - плотность распределения случайной ведичины $X$, оценивающей "принадлежность" объекта отрицательному классу.

Классифицирующая функция имеет вид $h_S(x, T) = sign(f(w, x) - T)$.

Вычислим ROC-AUC (площадь под кривой ROC):

$$
AUC 
= \int_0^1 f(x) \mathrm{d}x = [f(x) = TPR(T), x = FPR(T), T \in (-\infty, +\infty)]
= \int_{+\infty}^{-\infty} TPR(T) * FPR'(T) \mathrm{d}T
$$

Точке (0, 0) на графике ROC соответствует случай, когда классификатор считает, что все объекты принадлежат отрицаиельному классу $\rightarrow T = +\infty$. Точке (1, 1) на графике ROC соответствует случай, когда классификатор считает, что все объекты принадлежат положительному классу $\rightarrow T = -\infty$.

$$
AUC 
= \int_{+\infty}^{-\infty} TPR(T) * FPR'(T) \mathrm{d}T
= \int_{+\infty}^{-\infty} (-f_{neg}(T) \mathrm{d}T) \int_{T}^{+\infty} f_{pos}(T_0) \mathrm{d}T_0
= \int_{-\infty}^{+\infty} \int_{-\infty}^{+\infty} 1_{T0 > T} * f_{neg}(T) * f_{pos}(T_0) \mathrm{d}T \mathrm{d}T_0
= P(X_{pos} > X_{neg})
$$

Вероятность $P(X_{pos} > X_{neg})$ означает вероятность оценить объект положительного класса выше объекта отрицательного.

### Задание 2

**Presicion (точность)** показывает точность классификации объектов положительного класса, т.е. насколько много среди объектов, которые классификатор распознал как "положительные", объектов *true positive*.<br>
**Recall (полнота)** показывает как много объектов *true positive* по отношению ко всем возможным объектам положительного класса из распределения, то есть, грубо говоря, отношение количества положительно классифицированных объектов к количеству объектов, которые должны быть положительно классифицированы.

1) Presicion важнее Recall: Требуется получить как можно более точную классификацию, например, при фильтрации спама.

2) Recall важнее Presicion: Требуется получить как можно больше объектов положительного класса, например, при распознавании заболеваний по симптомам, определении мошенннических транзакций в выписке с банковского счета/карты.

3) Необходимо найти баланс. К примеру, при лексической нормализации коротких сообщений необходимо распознавать слова-кандидаты для нормализации максимально точно, но при этом не изменять слова, не требующие нормализации, то есть максимизировать нужно одновременно Precision и Recall.
